```python
# Lea la imagen de  que es una imagen a color
img_path = '/content/'
img_name = 'Color2.png'
medica_im = io.imread('Color2.png')
io.imshow(medica_im)
```

```python
# Imprimir informaci√≥n sobre la forma de la imagen y el tipo
print('La forma de la imagen es {} y el tipo {}'.format(medica_im.shape, medica_im.dtype))
```

```python
medica_im = io.imread(img_path.__add__(img_name), as_gray=True)
io.imshow(medica_im)
```

```python
tomografia_im = io.imread(img_path.__add__('tomografia.jpeg'))
tomografia_im
```

```python
# Convertir los datos en flotantes
from skimage import img_as_float 
tomografia_float = img_as_float(tomografia_im)
tomografia_float
```

```python
# Guardar imagen flotante
io.imsave(img_path.__add__('tomografia.jpeg'), tomografia_float)
```

```python
from skimage import exposure
bins, bins_c = exposure.histogram(tomografia_float)
```

```python
bins
```

```python
plt.plot(bins_c, bins)
```

```python
# Numpy library
import numpy as np
```

```python
bins, bins_e = np.histogram(tomografia_float, 255)
plt.plot(bins_e[1:], bins)
```
